**matplotlib and NumPy**

```python
import math
import numpy as np
t = np.arange(0,2.5,0.1)
print('t:\n',t)
y1 = np.sin(math.pi*t)
print('Y1:\n',+y1)
y2 = np.sin(math.pi*t+math.pi/2)
print('Y2:\n',+y2)
y3 = np.sin(math.pi*t-math.pi/2)
print('Y3:\n',+y3)
plt.plot(t,y1,'b*',t,y2,'g^',t,y3,'ys')
```

```python
plt.plot(t,y1,'b--',t,y2,'g',t,y3,'r-.')
```

**Working with Multiple Figures and Axes**

```python
t = np.arange(0,5,0.1)
y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)
```

```python
plt.plot(t,y1,'b-.',t,y2,'r--')
```

**To divide the canvas area into two horizontal areas**

subplot(nrows, ncols, index, **kwargs)

```python
plt.subplot(211)
plt.plot(t,y1,'b-.')
plt.subplot(212)
plt.plot(t,y2,'r--')
```

**To divide it vertically**

```python
t = np.arange(0.,1.,0.05)
y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)
plt.subplot(121)
plt.plot(t,y1,'b-.')
plt.subplot(122)
plt.plot(t,y2,'r--')
```

**To add text to the points in a graph**

```python
plt.axis([0,5,0,20])
plt.title('My first plot',fontsize=20,fontname='Times New Roman')
plt.xlabel('Counting',color='gray')
plt.ylabel('Square values',color='gray')
plt.text(1,1.5,'First') #add value above the point: plt.text(1,y+0.5,'First')
plt.text(2,4.5,'Second') # ADd value below the point: plt.text(1,y-0.5,'First')
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Fourth')
plt.plot([1,2,3,4],[1,4,9,16],'r^')
```

**To add a mathematical expression in the graph**

```python
plt.axis([0,5,0,20])
plt.title('My first plot',fontsize=20,fontname='Times New Roman')
plt.xlabel('Counting',color='gray')
plt.ylabel('Square values',color='gray')
plt.text(1,1.5,'First')
plt.text(2,4.5,'Second')
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Fourth')
plt.text(1.5,15,r'$y = x^2$',fontsize=20,bbox={'facecolor':'yellow',
'alpha':0.2})
plt.plot([1,2,3,4],[1,4,9,16],'r--.')
```

**Handling Date Values**

```python
import datetime
import numpy as np
import matplotlib.pyplot as plt
events = [datetime.date(2015,1,23),datetime.date(2015,1,28),datetime.date(2015,2,3),
          datetime.date(2015,2,21),datetime.date(2015,3,15),datetime.date(2015,3,24),
          datetime.date(2015,4,8),datetime.date(2015,4,24)]
readings = [12,22,25,20,18,15,17,14]
plt.plot(events,readings)
```

```python
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%Y-%m')
events = [datetime.date(2015,1,23),datetime.date(2015,1,28),
          datetime.date(2015,2,3),datetime.date(2015,2,21),
          datetime.date(2015,3,15),datetime.date(2015,3,24),
          datetime.date(2015,4,8),datetime.date(2015,4,24)]
readings = [12,22,25,20,18,15,17,14]
fig, ax = plt.subplots()
plt.plot(events,readings)
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
ax.xaxis.set_minor_locator(days)
```
